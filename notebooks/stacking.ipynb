# 2.Plot error distribution

```python
def smape(y_true, y_pred):
    
    # CONVERT TO NUMPY
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    
    # WHEN BOTH EQUAL ZERO, METRIC IS ZERO
    both = np.abs(y_true) + np.abs(y_pred)
    idx = np.where(both==0)[0]
    y_true[idx]=1; y_pred[idx]=1
    
    return 100/len(y_true) * np.sum(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))
```

```python
submission = submission.merge(train, on=["row_id"], how='left')
submission.head()
```

```python
cfips_list = train['cfips'].unique()
error_x = []
error_y = []
```

```python
cfips_list
```

```python
# calculate smape for each cfips

for cfips in tqdm(cfips_list):

    y_pred_x = submission.loc[
                (submission["cfips"] == cfips) & (submission["first_day_of_month"] == "2022-12-01")]["microbusiness_density_x"]
    
    y_pred_y = submission.loc[
                (submission["cfips"] == cfips) & (submission["first_day_of_month"] == "2022-12-01")]["microbusiness_density_y"]
    
    y_true = submission.loc[
                (submission["cfips"] == cfips) & (submission["first_day_of_month"] == "2022-12-01")]["microbusiness_density"]
    
    error_x.append(smape(y_pred_x, y_true))
    error_y.append(smape(y_pred_y, y_true))

error_x[:5]
```

```python
#plot smape for each cfips

fig = px.scatter(y=error_x, x=cfips_list, title='SMAPE for each cfips')
fig.show()
```
